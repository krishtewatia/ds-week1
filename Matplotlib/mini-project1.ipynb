```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('netflix_titles.csv')
df=df.dropna(subset=['type','release_year','rating','country','duration'])
type_counts=df['type'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_counts.index,type_counts.values,color=['skyblue','coral'])
plt.title('Number of movies vs TV shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('movies_vs_tvshows.png')
plt.show()
```

```python
rating_counts=df['rating'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(rating_counts,labels=rating_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Percentage of content creating')
plt.tight_layout()
plt.savefig('contentratings.png')
plt.show()
```

```python
movie_df=df[df['type']=='Movie'].copy()
movie_df['duration_int']=movie_df['duration'].str.replace('min','').astype(int)
plt.figure(figsize=(6,8))
plt.hist(movie_df['duration_int'],bins=30,color='purple',edgecolor='black')
plt.title('Distribution of movie duration')
plt.xlabel('Duration(minutes)')
plt.ylabel('Number of movies')
plt.tight_layout()
plt.savefig('movie_duration_histogram.png')
plt.show()
```

```python
release_counts=df['release_year'].value_counts().sort_index()
plt.figure(figsize=(10,6))
plt.scatter(release_counts.index,release_counts.values,color='red')
plt.title('Release year vs no. of shows')
plt.xlabel('Release Year')
plt.ylabel('Number of TV Shows')
plt.tight_layout()
plt.savefig('release_year.png')
plt.show()
```

```python
countr_counts=df['country'].value_counts().head(10)
plt.figure(figsize=(8,6))
plt.barh(countr_counts.index,countr_counts.values,color='teal')
plt.title('Top 10 countries by no. of shows')
plt.xlabel('No. of TV Shows')
plt.ylabel('Country')
plt.tight_layout()
plt.savefig('countr_Shows.png')
plt.show()
```
