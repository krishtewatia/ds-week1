```python
import matplotlib.pyplot as plt
```

```python
product=['A','B','C','D']
sales=[1000,1500,800,2000]
plt.barh(product,sales,color='orange',label='Sales 2025')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product sale comparison')
plt.legend()
plt.show()
```

```python
regions=['North','South','East','West']
revenue=[3000,2000,1500,1000]
plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors=['gold','coral','cyan','pink'])
plt.show()
```

```python
scores=[32,45,76,23,92,49,32,93,82,94,82,32,73,84,52,93,56,54,98,33]
plt.hist(scores,bins=5,color='purple',edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel('Number of Students')
plt.title('Score Distribution of Students')
plt.show()
```

```python
hours_studied=[1,2,3,4,5,6,7,8]
exams_score=[50,55,65,70,75,80,85,90]
plt.scatter(hours_studied,exams_score,color='green',marker='o',label='Student Data')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Scored')
plt.title('Relationship b/w study time and exam score')
plt.legend()
plt.grid()
plt.show()
```

```python

plt.scatter([1,2,3],[50,55,60],color='blue',label='Class A')
plt.scatter([1,2,3],[45,50,52],color='coral',label='Class B')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Scored')
plt.title('Comparison of two classes')
plt.legend()
plt.grid()
plt.show()
```

```python
x=[1,2,3,4]
y=[10,20,15,25]
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Line Chart')
plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('Bar Chart')
plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(10,5))
x=[1,2,3,4]
y=[10,20,15,25]
ax[0].plot(x,y)
ax[0].set_title('Line Plot')
ax[1].bar(x,y)
ax[1].set_title('Bar Chart')
fig.suptitle('Comparison of line and Bar Charts')
plt.show()
```

```python
x=[1,2,3,4]
y=[10,20,15,25]
plt.plot(x,y,color='blue',marker='o')
plt.title('Simple line graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.savefig('lineplot.png',dpi=300,bbox_inches='tight')
plt.show()
```
